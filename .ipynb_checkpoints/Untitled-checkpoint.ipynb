```python
dfq1 = card.loc[:,["Gender","Card_Category","Credit_Limit"]].groupby(["Gender","Card_Category"],as_index=False).quantile(0.25)
dfq1.columns = ["Gender","Card_Category","Credit_B1"]
dfq3 = card.loc[:,["Gender","Card_Category","Credit_Limit"]].groupby(["Gender","Card_Category"],as_index=False).quantile(0.75)
dfq3.columns = ["Gender","Card_Category","Credit_B3"]
dfq2 = card.loc[:,["Gender","Card_Category","Credit_Limit"]].groupby(["Gender","Card_Category"],as_index=False).quantile(0.50)
dfq2.columns = ["Gender","Card_Category","Credit_B2"]
dfq2
```

```python
df_iqr = dfq1.merge(dfq3, on = ["Gender", "Card_Category"], how = "inner").merge(dfq2, on = ["Gender", "Card_Category"], how = "inner")
df_iqr["IQR"] = df_iqr.apply(lambda r: r["Credit_B3"] - r["Credit_B1"], axis=1)
df_iqr["Upper"] = df_iqr.apply(lambda j: (j["Credit_B3"] + 1.5*j["IQR"]), axis=1)
df_iqr["Lower"] = df_iqr.apply(lambda j: (j["Credit_B1"] - 1.5*j["IQR"]), axis=1)
df_iqr
```

```python
card_t = card.merge(df_iqr, on = ["Gender", "Card_Category"], how = "left").copy()
card_t.shape
```

```python
def getCredit(r):
    if ((r["Credit_Limit"] <= r["Lower"]) or (r["Credit_Limit"]>= r["Upper"])):
        return r["Credit_B2"]
    else:
        return r["Credit_Limit"]

    
    
card_t["New_Credit"] = card_t.apply(lambda r: getCredit(r), axis = 1)
card_t
```

```python
sns.boxplot(data = card_t.loc[ (card_t['Gender'] == 'M'), : ], x= 'Card_Category', y ='New_Credit' )
```

```python
cols = ['Attrition_Flag', 'Gender', 'Education_Level', 'Marital_Status',
       'Income_Category', 'Card_Category']

valids = []
many = []
one = []

size = card.shape[0]
umbral = 0.9

for c in cols : 
    level = card.loc[:,cols].drop_duplicates().shape[0]
    if (level/size) > umbral: 
        many.append(c)
    elif level == 1: 
        one.append(c)
    else: 
        valids.append(c)
        
valids
```

```python
level = card.groupby(["Education_Level"], as_index = False).agg({"Customer_Age":["count"]})
level.columns = level.columns.droplevel(1)
level
```

```python
level["Percent"] = level["Customer_Age"]/size
level.sort_values(["Percent"], ascending = False)
```

```python
level["Education_t"] = "Others"
level["Education_t"] = level.apply(lambda r: r["Education_Level"] if r["Percent"] > 0.1 else r["Education_t"], axis=1)
level.loc[:,["Education_Level", "Education_t"]]
```

```python
card = card.merge(level, on= ["Education_Level"], how = "inner")
card
```
